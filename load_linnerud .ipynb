```python
import pandas as pd
import numpy as np

import seaborn as sns
from sklearn.datasets import load_linnerud
```

```python
load_data = load_linnerud()

load_data
```

```python
#DATAFrame
data_set = pd.DataFrame(load_data.data)
```

```python
data_set.columns = load_data.feature_names
```

```python
data_set
```

```python
#Dependent and Independent x , y
x = data_set
y = load_data.target
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
#Standardazation

from sklearn.preprocessing import StandardScaler

Standardscalr = StandardScaler()

Standardscalr
```

```python
Standardscalr.fit_transform(X_train , y_train)
```

```python
Standardscalr.transform(X_test)
```

```python
##LineraImplemation

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
linear_reg = LinearRegression()
linear_reg
```

```python
linear_reg.fit(X_train , y_train)
```

```python
score = cross_val_score( linear_reg , X_train , y_train , scoring='neg_mean_squared_error' , cv=5)
```

```python
score
```

```python
np.mean(score)
```

```python
x_pred = linear_reg.predict(X_test)
```

```python
x_pred
```

```python
import matplotlib.pyplot as plt
```

```python

sns.displot(x_pred - y_test, kind='kde', fill=True)
plt.title('KDE of Predictions vs Actuals')
plt.xlabel('Prediction Error')
plt.ylabel('Density')


plt.figure(figsize=(8, 6))
plt.scatter(y_test, x_pred ,  color='blue', label='Predicted vs Actual')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linewidth=2, label='Best Fit Line')
plt.title('Linear Regression Best Fit Line')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score
```
